```python
# check the summary statistics for each of the crops

from ipywidgets import interact

def print_statistics(series, label):
      #"""Helper function to print statistics for a given Series and label."""
    if not series.empty:
        print(f"Statistics for {label}:")
        print("Minimum required:", series.min())
        print("Average required:", series.mean())
        print("Maximum required:", series.max())
    else:
        print(f"No data available for {label}.")

@interact
def summary(crop=list(data['label'].value_counts().index)):
    x = data[data['label'] == crop]

    print(f"\nStatistics for Nitrogen for {crop}:")
    print_statistics(x['N'], "Nitrogen")

    print(f"\nStatistics for Phosphorus for {crop}:")
    print_statistics(x['P'], "Phosphorus")

    print(f"\nStatistics for Potassium for {crop}:")
    print_statistics(x['K'], "Potassium")

    print(f"\nStatistics for Temperature for {crop}:")
    print_statistics(x['temperature'], "Temperature")

    print(f"\nStatistics for Humidity for {crop}:")
    print_statistics(x['humidity'], "Humidity")

    print(f"\nStatistics for PH for {crop}:")
    print_statistics(x['ph'], "PH")

    print(f"\nStatistics for Rainfall for {crop}:")
    print_statistics(x['rainfall'], "Rainfall")
```

```python
from sklearn.cluster import KMeans
#removing the labels colum
x = data.drop(['label'], axis = 1)
#selecting all the values of the data
x = x.values
#checking shape
print(x.shape)
```

```python

# Example dataset (replace with your own data, 'x' must be defined)
# x = your_data_array 
# Set the figure size
plt.figure(figsize=(10, 4))

# Initialize WCSS (Within-Cluster-Sum-of-Squares) list
wcss = []

# Loop through 1 to 10 clusters
for i in range(1, 11):
    km = KMeans(n_clusters=i, init="k-means++", max_iter=300, n_init=10, random_state=0)
    km.fit(x)  # Fit the model to the data
    wcss.append(km.inertia_)  # Append the inertia (WCSS) to the list

# Plot the Elbow graph
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='b', markersize=8)
plt.title('Elbow Method For Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))  # Ensure x-ticks match the number of clusters
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
#implement the k means algorithm to perform clustering analysis
km = KMeans(n_clusters=4, init="k-means++", max_iter=300, n_init=10, random_state=0)
y_means =    km.fit_predict(x)
#find out the results
a = data['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis = 1)
z = z.rename(columns = {0: 'cluster'})

#checking cluster of each crops
print("lets check the results After Appling the K Means Clustering Analysis \n")
print("Crops in First Cluster : ", z[z['cluster'] ==0]['label'].unique())
print(".............................................................")
print("Crops in Second Cluster : ", z[z['cluster'] ==1]['label'].unique())
print(".............................................................")
print("Crops in Third Cluster : ", z[z['cluster'] ==2]['label'].unique())
print(".............................................................")
print("Crops in Four Cluster : ", z[z['cluster'] ==3]['label'].unique())
```
